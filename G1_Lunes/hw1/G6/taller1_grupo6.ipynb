```python
import matplotlib.pyplot as plt
import numpy as np

class Escena:
    def __init__(self, fondo_color="blue", punto_color="red", punto_posicion=np.array([0, 0]), fondo_dimensiones=(10, 10)):
        # Definimos el color del fondo, el color del punto, la posición del punto (como un vector) y las dimensiones del fondo
        self.fondo_color = fondo_color
        self.punto_color = punto_color
        self.punto_posicion = punto_posicion  # Usamos un vector para la posición del punto
        self.fondo_dimensiones = fondo_dimensiones

    def mostrar(self):
        # Crear una figura y un eje para dibujar la escena
        fig, ax = plt.subplots()

        # Dibujo del fondo (rectángulo)
        ax.add_patch(plt.Rectangle((0, 0), self.fondo_dimensiones[0], self.fondo_dimensiones[1],
                                   color=self.fondo_color))

        # Dibujo del punto (si está dentro del área visible)
        ax.plot(self.punto_posicion[0], self.punto_posicion[1], 'o', color=self.punto_color)

        # Ajustamos límites para que todo sea visible
        ax.set_xlim(0, self.fondo_dimensiones[0])
        ax.set_ylim(0, self.fondo_dimensiones[1])
        ax.set_aspect('equal', 'box')

        # Mostrar la escena
        plt.show()

    def desplazar_punto(self, vector_desplazamiento):
        """
        Desplaza el punto usando un vector de desplazamiento.
        El vector de desplazamiento también será un array numpy.
        """
        self.punto_posicion += vector_desplazamiento  # Realizamos la operación vectorial para desplazar el punto

    def rotar_punto(self, angulo):
        """
        Rota el punto alrededor del origen (0, 0) por un ángulo dado.
        El ángulo se proporciona en grados.
        """
        # Convertir el ángulo a radianes
        angulo_rad = np.deg2rad(angulo)

        # Matriz de rotación 2D
        matriz_rotacion = np.array([[np.cos(angulo_rad), -np.sin(angulo_rad)],
                                    [np.sin(angulo_rad), np.cos(angulo_rad)]])

        # Realizar la rotación
        self.punto_posicion = np.dot(matriz_rotacion, self.punto_posicion)

# Ejemplo de uso con tres escenas

# Primera escena
escena1 = Escena(fondo_color="lightgreen", punto_color="purple", punto_posicion=np.array([5, 5]), fondo_dimensiones=(15, 15))
escena1.mostrar()

# Segunda escena
escena2 = Escena(fondo_color="skyblue", punto_color="orange", punto_posicion=np.array([3, 7]), fondo_dimensiones=(15, 15))
escena2.mostrar()

# Tercera escena
escena3 = Escena(fondo_color="yellow", punto_color="black", punto_posicion=np.array([2, 2]), fondo_dimensiones=(15, 15))
escena3.mostrar()

# Operaciones vectoriales sobre las escenas
# Desplazamos el punto de la escena 1 por el vector [2, 3]
escena1.desplazar_punto(np.array([2, 3]))
escena1.mostrar()

# Rotamos el punto de la escena 2 20 grados
escena2.rotar_punto(20)
escena2.mostrar()

# Desplazamos el punto de la escena 3 por el vector [-1, -1]
escena3.desplazar_punto(np.array([-1, -1]))
escena3.mostrar()
```
