```python
import numpy as np
from sympy import*
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
init_printing()
```

```python
def dibujar_punto(punto, label, x_lim = [0, 1], y_lim=[0, 1], z_lim=[0, 1]):
  """
  Dibuja un punto en un espacio tridimensional.

  Parametros:
    punto: Un vector NumPy de tres elementos representando las coordenadas (x, y, z) del punto.
    label: Una cadena de texto para etiquetar el punto en el gráfico.
    x_lim: Una lista de dos elementos que define los límites del eje x.
    y_lim: Una lista de dos elementos que define los límites del eje y.
    z_lim: Una lista de dos elementos que define los límites del eje z.
  """

  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  ax.scatter(punto[0], punto[1], punto[2], color='b', label=label)

  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')
  ax.set_title('Espacio 3D')

  ax.quiver(0, 0, 0, punto[0], punto[1], punto[2], color='b')

  ax.set_xlim(x_lim)
  ax.set_ylim(y_lim)
  ax.set_zlim(z_lim)
  ax.legend()
  plt.show()
```

```python
## Angulo simbolico
theta = symbols("theta",  real=True)
```

```python
## Matrix de rotación simbolica sobre le eje X
R_x = Matrix([[1, 0, 0], [0, cos(theta), -sin(theta)], [0, sin(theta), cos(theta)]])
R_x
```

```python
P0 = np.array([1, 1, 0])
print(P0)
```

```python
dibujar_punto(P0, "Punto P0", y_lim=[-1,1])
```

```python
theta_val = np.pi #180 grados en radianes
R_p0 = np.array(R_x.subs({theta: theta_val})).astype(np.float64) #Se convierte theta simbolico a un valor concreto

P0_rotado = np.dot(R_p0, P0) #Producto punto entre la matrix de rotación y el vector de posición del punto
print(P0_rotado)
```

```python
dibujar_punto(P0_rotado, "Punto P0 rotado", y_lim=[-1, 1])
```

```python
dibujar_punto(P0_rotado, "Punto P0 rotado", y_lim=[0, 1])
```

```python
R_y = Matrix([[cos(theta), 0, sin(theta)], [0, 1, 0], [-sin(theta), 0, cos(theta)]])
R_y
```

```python
R_z = Matrix([[cos(theta), -sin(theta), 0], [sin(theta), cos(theta), 0], [0, 0, 1]])
R_z
```

```python
def trasladar_punto(punto, vector_traslacion):
    """
    Realiza la traslación de un punto en el espacio tridimensional.

    Parámetros:
      punto: Un vector  de tres elementos con las coordenadas (x, y, z) del punto.
      vector_traslacion: Traslacion de un vector de tres elementos con las coordenadas (x, y, z) del punto.

    """

    return np.add(punto, vector_traslacion)

    """
        Retorna el punto trasladado.
    """
# Ejemplo de uso
punto = np.array([2, 4, 7])
vector_traslacion = np.array([2, 3, 1])

punto_trasladado = trasladar_punto(punto, vector_traslacion)

print("Punto original:", punto)
print("Vector de traslación:", vector_traslacion)
print("Punto trasladado:", punto_trasladado)
```

```python
#Dibuja los puntos
figura = plt.figure()
dibujo = figura.add_subplot(111, projection='3d')

#Dibuja el punto original
dibujo.scatter(punto[0], punto[1], punto[2], color='blue', label='Punto original')
#Dibuja el punto de Traslado
dibujo.scatter(punto_trasladado[0], punto_trasladado[1], punto_trasladado[2], color='red', label='Punto trasladado')

#Recta que comunica los puntos
dibujo.plot([punto[0], punto_trasladado[0]],
                [punto[1], punto_trasladado[1]],
                [punto[2], punto_trasladado[2]], color='green', linestyle='--')

#Ejes
dibujo.set_xlabel('X')
dibujo.set_ylabel('Y')
dibujo.set_zlabel('Z')
plt.legend()
plt.title('Traslación en el espacio 3D')
plt.show()
```
